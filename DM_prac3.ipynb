```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import math
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
data = pd.read_csv('xAPI-Edu-Data.csv')
```

```python
data1=data.head(100)
```

```python
data1
```

```python
data1=data1.sort_values('raisedhands', ascending=[True])
```

```python
data1
```

```python
df1=data1[['raisedhands']]
```

```python
df1
```

```python
df1['index'] = range(0,len(df1))
df1.set_index('index', inplace=True)
```

```python
df1
```

```python
print(len(df1))
```

```python
dfmean=df1
```

```python
for i in range(0,len(dfmean),5):
    mean = np.mean(dfmean.iloc[i:i+5])
    if(i+5<=len(dfmean)):
        k=i+5
    else:
        k=len(dfmean)
    for j in range(i,k):
        dfmean.iloc[j] = mean
```

```python
dfmean
```

```python
dfmedian=data1[['raisedhands']]
```

```python
dfmedian
```

```python
dfmedian['index'] = range(0,len(dfmedian))
dfmedian.set_index('index', inplace=True)
```

```python
dfmedian
```

```python
for i in range(0,len(dfmedian),5):
    med = np.median(dfmedian.iloc[i:i+5])
    if(i+5<=len(dfmedian)):
        k=i+5
    else:
        k=len(dfmedian)
    for j in range(i,k):
        dfmedian.iloc[j] = med
```

```python
dfmedian
```

```python
dfbound=data1[['raisedhands']]
```

```python
dfbound['index'] = range(0,len(dfbound))
dfbound.set_index('index', inplace=True)
```

```python
dfbound
```

```python
for i in range(0,len(dfbound),5):
    minimum = np.min(dfbound.iloc[i:i+5])
    maximum= np.max(dfbound.iloc[i:i+5])
    if(i+5<=len(dfbound)):
        k=i+5
    else:
        k=len(dfbound)
    for j in range(i,k):
        diff1=dfbound.iloc[j] - minimum
        diff2=maximum - dfbound.iloc[j]
        if int(diff1)<int(diff2):
            dfbound.iloc[j]=minimum
        else:
            dfbound.iloc[j]=maximum
       
```
