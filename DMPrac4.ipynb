```python
#importing libraries

import numpy as np
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def MinMax(data):
    newMax = 1
    newMin = 0
    maxValue = np.max(data)
    minValue = np.min(data)
    normalizedValues = list()
    
    print("Max value:",maxValue,"\nMin Value:",minValue)
    for i in data:
        temp = ((i - minValue)/(maxValue - minValue))*(newMax - newMin) + newMin
        normalizedValues.append(temp)
        
    return normalizedValues
```

```python
def zScore(data):
    mean = np.mean(data)
    sd = np.std(data)
    
    print("\nMean of the data:",mean,"\nStandard deviation of the data",sd)
    normalizedValues = list()
    
    for i in data:
        temp = (i - mean)/sd
        normalizedValues.append(temp)
        
    return normalizedValues
```

```python
def decimalScaling(data):
    maxno = np.max(data)
    j = len(str(abs(maxno)))
    
    print(j)
    normalizedValues = list()
    for i in data:
        temp = i/(10**j)
        normalizedValues.append(temp)
        
    return normalizedValues
```

```python
data = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 33, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]

print("Min Max normalization:")
resultMM = MinMax(data)
print("\nAfter Min Max normalization:\n",resultMM)
print("for 25:",resultMM[10])
print("for 52:",resultMM[-2])
```

```python
print("\n\nZ-score normalization:")
resultZS = zScore(data)
print("\nAfter z-score normalization:\n",resultZS)
print("for 35:",resultZS[-7])
```

```python
print("\n\nDecimal Scaling normalization:")
resultDS = decimalScaling(data)
print("\nAfter decimal scaling normalization:\n",resultDS)
print("for 35:",resultDS[-7])
```

# Conclusion

Similarly, the goal of normalization is to change the values of numeric columns in the dataset to a common scale, without distorting differences in the ranges of values. For machine learning, every dataset does not require normalization. It is required only when features have different ranges.Normalization is a good technique to use when you do not know the distribution of your data or when you know the distribution is not Gaussian (a bell curve)

Using normalization, we have less biased dataset, using z-score normalization we overcame the outliers and for columns like longitute and latitude the ranges where brought into 0 and 1
