```python
import pandas as pd

lst = [['T10','Beer Nuts Diaper'],['T20','Beer Coffee Diaper'],['T30','Beer Diaper Eggs'],['T40','Nuts Eggs Milk'],['T50','Nuts Coffee Diaper Eggs Milk']]
df = pd.DataFrame(lst, columns =['Tid', 'Items Bought']) 
print(df) 
```

```python
d=set()
for i in df['Items Bought']:
    j=i.split()
    for x in j:
        d.add(x)
              
print(d)
```

```python
s = 0.4
c = 0.4
items = [i.split(' ') for i in df['Items Bought']]
items
```

```python
value ={i:0 for i in d}
for i in items:
    for j in i:
        value[j]+=1
value
```

```python
def calc_support(its,idlst):
    its = set(its)
    support = 0
    for cur_list in idlst:
        if len(its.intersection(set(cur_list))) == len(its):
            support+=1
    return support/len(idlst)
```

```python
calc_support(['Eggs'],items)
```

```python
def mineRules(p,q,final):
    global items
    if len(p):
        val = calc_support(p,items)
        if val>=s:
            final[' '.join(p)] = val
    if len(q)==0:    
        return 
    if len(p)==0:
        mineRules(p+[q[0]],q[1:],final)
        mineRules(p,q[1:],final)
    else:
        if val>=s:
            final[' '.join(p)] = val
            mineRules(p+[q[0]],q[1:],final)
            mineRules(p,q[1:],final)
        return
```

```python
final = {}
p,q = [],list(d)
mineRules(p,q,final)
final
```

```python
import numpy as np
#mining Rules
rules=list(final.keys())
le=[]
for i in rules:
    le.append(len(i.split()))
    
m=max(le)
print(m)

word=[]
for i in rules:
    j=i.split()
    if(len(j)==m):
        word.append(j)
        
main=list(np.concatenate(word).flat)
print(main)
for i in rules:
    if(len(i.split())==m):
        confi=final[i]
print(confi)
```
